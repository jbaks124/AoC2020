```python
#Day3
# part 1

import urllib
f = open("input.txt", "r")
lines = f.read()
values = lines.splitlines()
```

```python
for i in values[0:5]:
    print(i)
```

```python
data = {}
linecount = 0


for x in values:
    
    prepeat = ""
    
    c = 0
    while c < 75:
        prepeat += x
        c += 1
    
    data[linecount] = {"p" : prepeat}
    linecount += 1
```

```python
t = 100
print(data[t]['p'])     
print(len(data[t]['p']))
```

```python
output = {}
pos = 1
angle = 3
trees = 0
indicator = ""

for x in data:
    if data[x]['p'][pos-1:pos] == '#':
        trees += 1
        indicator = "X"
    else:
        indicator = "O"   
        
    a = data[x]['p'][:pos-1]
    b = data[x]['p'][pos+1:]
    c = a + indicator + b   
    output[x] = {"p" : c}  
    pos += angle
    
print('encountered trees:' + str(trees))
print('required width:' + str(pos))
```

```python
#part 2

output2 = {}
traversetrees = {}
traverse = {}
traverse[0] = {"right": 1, "down": 1}
traverse[1] = {"right": 3, "down": 1}
traverse[2] = {"right": 5, "down": 1}
traverse[3] = {"right": 7, "down": 1}
traverse[4] = {"right": 1, "down": 2}

for i in traverse:
    #print(traverse[i])
    pos = 1
    angle = traverse[i]['right']
    speed = traverse[i]['down']
    trees = 0
    trajection = 1

    for x in data:
        trajection -= 1
        if trajection == 0:
            trajection = speed
            if data[x]['p'][pos-1:pos] == '#':
                trees += 1
                indicator = "X"
            else:
                indicator = "O"   

            a = data[x]['p'][:pos-1]
            b = data[x]['p'][pos+1:]
            c = a + indicator + b   
            output2[i,x] = {"p" : c}
            pos += angle
        
    traversetrees[i] = trees

print(traversetrees)

outcome = traversetrees[0]*traversetrees[1]*traversetrees[2]*traversetrees[3]*traversetrees[4]
print(outcome)
```
