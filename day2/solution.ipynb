```python
#Day 2

import urllib
f = open("input.txt", "r")
lines = f.read()
values = lines.splitlines()
```

```python
for i in values[0:5]:
    print(i)
```

```python
data = {}
linecount = 0

for x in values:
    arr = x.split(' ')
    arr2 = arr[0].split('-')
    min = int(arr2[0])
    max = int(arr2[1])
    letter = arr[1][0:1]
    password = arr[2]
    data[linecount] = {"min" : min, "max" : max, "letter" : letter, "password" : password}
    linecount += 1
```

```python
for x in data:
    print(data[x])
```

```python
#1
valid = 0

for x in data:
    count = data[x]['password'].count(data[x]['letter'])
    if count >= data[x]['min'] and count <= data[x]['max']:
        valid += 1

print(valid)
    
```

```python
#2
datav = {}
valid = 0

for x in data:
    min = data[x]['min']
    max = data[x]['max']
    if data[x]['letter'] == data[x]['password'][min-1:min] or data[x]['letter'] == data[x]['password'][max-1:max]:
        valid += 1
        datav[x] = 'yes'
    else:
        datav[x] = 'no'

    # Almost forgot ;-)
    if data[x]['letter'] == data[x]['password'][min-1:min] and data[x]['letter'] == data[x]['password'][max-1:max]:
        valid -= 1

        
print(valid)
```

```python
for x in datav:
    print(x, datav[x])
```
